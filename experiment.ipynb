```python
import os
import json
import csv
from datasets.loader import load_multiarith, load_gsm8k
from pipeline import ZeroShotCOTPipeline, DirectAnswerPipeline
from evaluate import evaluate
from gemini_client import GeminiClient
```

```python
def save_results(results, out_dir="./outputs", filename="results"):
    os.makedirs(out_dir, exist_ok=True)

    # Save JSON
    json_file = os.path.join(out_dir, f"{filename}.json")
    with open(json_file, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

   

    print(f"Results saved to:\n  {json_file}\n")
```

```python
API_KEY = os.getenv("GEMINI_API_KEY")

def main():
    client = GeminiClient(API_KEY)
    pipeline = DirectAnswerPipeline(client)

    dataset = load_gsm8k("./dataset/grade-school-math/test.jsonl", limit=50)

    results = [pipeline.run_sample(sample) for sample in dataset]

    save_results(results, out_dir="./outputs", filename="gsmk_zero_shot")


if __name__ == "__main__":
    main()
```
